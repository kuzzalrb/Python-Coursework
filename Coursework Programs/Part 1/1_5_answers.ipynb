```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('student.csv')
data.head(3)
```

```python
plt.pie(data.groupby('school')['school'].count(),
labels=['Gabriel Pereira', 'Mousinho da Silveira'],
colors=["#66cc99", "#ffff66"],
autopct="%.1f%%",
)
plt.title("Pie-chart representing number of Students by School")
```

```python
maleStudents = data[data['sex']== 'M'].groupby('school')['sex'].count()
maleStudents
```

```python
dictMaleStudents = {}
keyCount = 0
for i in maleStudents.index:
    dictMaleStudents[keyCount] = maleStudents[i]
    keyCount += 1
dictMaleStudents
```

```python
femaleStudents = data[data['sex']== 'F'].groupby('school')['sex'].count()
femaleStudents
```

```python
dictFemaleStudents = {}
keyCount = 0
for i in femaleStudents.index:
    dictFemaleStudents[keyCount] = femaleStudents[i]
    keyCount += 1
dictFemaleStudents
```

```python
plt.figure(figsize=(10,6))
plt.bar(np.array(list(dictMaleStudents.keys())),
np.array(list(dictMaleStudents.values())),
color="#3399ff",
width=0.25,
label = 'Male'
)

plt.bar(np.array(list(dictFemaleStudents.keys()))+0.25,
np.array(list(dictFemaleStudents.values())),
color="#ffccff",
width=0.25,
label = 'Female'
)
plt.title("Bar graph of total number of male and female students by school")
plt.xlabel("School")
plt.ylabel ("Number of Students")
plt.xticks (np.array(list (dictMaleStudents.keys ())).astype("int64" )+0.12,
['Gabriel Pereira', 'Mousinho da Silveira'])
plt.legend()
```
