```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_horse_asses = pd.read_csv('horseasses-population-in-nepal-by-district.csv')
data_milk = pd.read_csv('milk-animals-and-milk-production-in-nepal-by-district.csv')
data_meat = pd.read_csv('net-meat-production-in-nepal-by-district.csv')
data_cotton = pd.read_csv('production-of-cotton-in-nepal-by-district.csv')
data_egg = pd.read_csv('production-of-egg-in-nepal-by-district.csv')
data_rabbit = pd.read_csv('rabbit-population-in-nepal-by-district.csv')
data_wool = pd.read_csv('wool-production-in-nepal-by-district.csv')
data_yak = pd.read_csv('yak-nak-chauri-population-in-nepal-by-district.csv')
```

```python
data_horse_asses['DISTRICT'] = data_horse_asses['DISTRICT'].str.upper()
data_milk['DISTRICT'] = data_milk['DISTRICT'].str.upper()
data_meat['DISTRICT'] = data_meat['DISTRICT'].str.upper()
data_cotton['DISTRICT'] = data_cotton['DISTRICT'].str.upper()
data_egg['DISTRICT'] = data_egg['DISTRICT'].str.upper()
data_rabbit['DISTRICT'] = data_rabbit['DISTRICT'].str.upper()
data_wool['DISTRICT'] = data_wool['DISTRICT'].str.upper()
data_yak['DISTRICT'] = data_yak['DISTRICT'].str.upper()
```

#### MERGE

```python
data_cotton_horse = data_cotton.merge(data_horse_asses, on = 'DISTRICT', how = 'outer')
data_cotton_horse_milk = data_cotton_horse.merge(data_milk, on = 'DISTRICT', how = 'outer')
data_cotton_horse_milk_meat = data_cotton_horse_milk.merge(data_meat, on = 'DISTRICT', how = 'outer')
data_cotton_horse_milk_meat_egg = data_cotton_horse_milk_meat.merge(data_egg, on = 'DISTRICT', how = 'outer')
data_cotton_horse_milk_meat_egg_rabbit = data_cotton_horse_milk_meat_egg.merge(data_rabbit, on = 'DISTRICT', how = 'outer')
data_cotton_horse_milk_meat_egg_rabbit_wool = data_cotton_horse_milk_meat_egg_rabbit.merge(data_wool, on = 'DISTRICT', how = 'outer')
data_merge = data_cotton_horse_milk_meat_egg_rabbit_wool.merge(data_yak, on = 'DISTRICT', how = 'outer')
data_merge
```

#### CLEAN

```python
'''Convert all district names to uppercase'''
data_merge['DISTRICT'] = data_merge['DISTRICT'].str.upper()

'''delete a row containing DISTRICT as Total'''
data_merge.loc[data_merge['DISTRICT'] == 'TOTAL']
data_merge = data_merge.drop(data_merge.index[6])

'''delete a row containing DISTRICT as Nepal'''
data_merge.loc[data_merge['DISTRICT'] == 'NEPAL']
data_merge = data_merge.drop(data_merge.index[3])

'''Drop the columns containing sum of other columns'''
data_merge = data_merge.drop(['TOTAL MILK PRODUCED', 'TOTAL MEAT', 'TOTAL EGG',], axis= 1)

data_merge.head(50)
```

```python
prod_array = data_merge['PROD. (Mt.)'].head(3).values
print(prod_array)

dist_array = data_merge['DISTRICT'].head(3).values
print(dist_array)
```

```python
plt.pie(prod_array,
labels=dist_array,
colors=["#66cc99", "#ffff66", "#66b3ff"],
autopct="%.1f%%",
)
plt.title("Pie-chart representing the percentage of cotton production by district")
```

```python
'''creating a dictionary with district names as keys and the number of NaNs as values'''
dictDistrictNullCount = {}
for i in range(len(data_merge['DISTRICT'])):
    district = data_merge['DISTRICT'].iloc[i]
    nullCount = str(data_merge.iloc[i,:].isnull().sum())
    dictDistrictNullCount[district] = nullCount
```

```python
dictDistrictNullCount
```

```python
data_items = dictDistrictNullCount.items()
data_list = list(data_items)

data_district_diversity = pd.DataFrame(data_list, columns=['District', 'Null columns'])
data_district_diversity = data_district_diversity.sort_values(by='District', ascending=True)
data_district_diversity.head(50)
```

```python
data_district_diversity['Null columns'].unique()
```

```python

sns.catplot(x="Null columns", y="District", data=data_district_diversity, height=20, aspect=15/20)
```
